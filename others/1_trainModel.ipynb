## 1. 处理从ChEMBL下载的数据

```python
import pandas as pd
df = pd.read_csv('AChE.Ki.csv',sep=';')  #ChEMBL上下载的csv是以分号为分隔符的
df.columns   # ChEMBL给出了41列数据
```

```python
df # 初始数据内容
```

```python
names = ['Molecule ChEMBL ID', 'Molecule Name', 'Smiles', 
         'Standard Type', 'Standard Relation', 'Standard Value', 'Standard Units',
         'Data Validity Comment', 'Comment',] # 仅关注这几列数据
df = df[names]
df
```

```python
# 查看数据中comment列的内容
df1 = df[df['Data Validity Comment'].notnull()]
df2 = df[df['Comment'].notnull()]
df1  #查看 Data Validity Comment内容
```

```python
df2 #查看comment内容
```

```python
# 查看数据注释发现，有数据单位不标准的，有Ki与IC50结论相反的。鉴于数据量足够多，可直接把这些有问题的数据舍去。
selection1 = df['Data Validity Comment'].isna()
selection2 = []
for comment in df['Comment']:
    keep = True
    if comment and isinstance(comment,str) and comment.find('corresponding IC50 reported as Active') > 0: keep = False
    selection2.append(keep)
selection3 = df['Smiles'].notnull() # Smiles数据不能为空
df = df[selection1 & selection2 & selection3]
df
```

```python
# Ki < 10000 nM的数据标记为active，剩余的标记为inactive
active_selection = (df['Standard Relation']=="\'=\'") & (df['Standard Value'] < 1e4)
active = df[active_selection]
active
```

```python
inactive = df[~active_selection]
inactive
```

## 2. 计算分子描述符

```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole # 在jupyter-notebook中显示分子结构
from mordred import Calculator, descriptors
from molvs import Standardizer
import pickle as pickle
import os
```

```python
def calc_descriptors(SMI_df):
    # create descriptor calculator with all descriptors
    calc = Calculator(descriptors, ignore_3D=True)
    s = Standardizer()
    mols = [Chem.MolFromSmiles(smi) for smi in SMI_df]
    mols = [s.largest_fragment(m) for m in mols]
    df = calc.pandas(mols)
    return df
# --- I/O function
def save(data,f):
    with open(f,"w") as f1:
        pickle.dump(data,f1)
    
def load(f):
    with open(f,"r") as f1:
        d = pickle.load(f1)
    return d
```

```python
#计算active(433个分子)和inactive(1147个分子)分别用时4分钟和8分钟。计算完成后用pickle将其保存，这样下次运行时节省时间。
#（若active和inactive内容有变则需删除临时文件再运行）
if os.path.isfile('descriptors-ki.data') :
    active_descriptors,inactive_descriptors = load('descriptors-ki.data')
else:
    active_descriptors = calc_descriptors(active['Smiles'])
    inactive_descriptors = calc_descriptors(inactive['Smiles'])
    save((active_descriptors,inactive_descriptors),'descriptors-ki.data')
```

```python
active_descriptors['labels']=1
inactive_descriptors['labels']=0
descriptors_ki = pd.concat([active_descriptors,inactive_descriptors],ignore_index=True)
descriptors_ki = descriptors_ki.select_dtypes(include='number')
descriptors_ki.shape
```

## 3. 训练预测模型

```python
from sklearn import svm 
from sklearn import preprocessing 
from sklearn.model_selection import train_test_split

clf = svm.SVC(gamma='scale')  #定义分类器classifier。(合理使用SVM需要调参确定C和gamma，这里不做展开，仅用默认参数)

features = descriptors_ki.columns[:-1]
y = descriptors_ki['labels'].values
X = descriptors_ki[features].values

scaler = preprocessing.StandardScaler().fit(X)
X = scaler.fit_transform(X)  #将数据标准化 

X_train, X_test, y_train, y_test = train_test_split(X,y) # 分割为训练集（3/4）和测试集(1/4)，
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
accuracy
```

```python
save((clf,scaler,features),'model')
```
